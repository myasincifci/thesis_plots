```python
import torch
from flow_matching.path.scheduler import CondOTScheduler, VPScheduler

from matplotlib import pyplot as plt
```

```python
t = torch.linspace(0, 1, steps=100)

ot_schedule = CondOTScheduler()(t)
vp_schedule = VPScheduler()(t)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

axes[0].plot(t.numpy(), ot_schedule.alpha_t.numpy(), label='alpha_t')
axes[0].plot(t.numpy(), ot_schedule.sigma_t.numpy(), label='sigma_t')
axes[0].plot(t.numpy(), torch.sqrt(ot_schedule.alpha_t**2 + ot_schedule.sigma_t**2).numpy(), label='sqrt(alpha_t + sigma_t)')
axes[0].set_title('OT Schedule')
axes[0].set_xlabel('t')
axes[0].set_ylabel('value')
axes[0].legend()

axes[1].plot(t.numpy(), vp_schedule.alpha_t.numpy(), label='alpha_t')
axes[1].plot(t.numpy(), vp_schedule.sigma_t.numpy(), label='sigma_t')
axes[1].plot(t.numpy(), torch.sqrt(vp_schedule.alpha_t**2 + vp_schedule.sigma_t**2).numpy(), label='sqrt(alpha_t + sigma_t)')
axes[1].set_title('VP Schedule')
axes[1].set_xlabel('t')
axes[1].legend()

plt.tight_layout()
plt.show()
```
